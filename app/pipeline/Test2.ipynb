```python
import pandas as pd
from sqlalchemy import create_engine

dtypes = {
    "LocationID": "int64",
    "Borough": "str",
    "Zone": "str",
    "service_zone": "str"
}
chunksize = 10000
df = pd.read_csv(
    'taxi_zone_lookup.csv',
    dtype=dtypes,
    iterator=True,
    chunksize=chunksize
)

# print(df)
engine = create_engine(f'postgresql+psycopg://root:root@localhost:5433/aegley_db')
first = True
for df in df:
    if first == True:
        df.to_sql(name='taxi_zone_lookup', con=engine, if_exists="replace")
    first = False
    df.to_sql(name='taxi_zone_lookup', con=engine, if_exists='append')
# for col, dtype in df.dtypes.items():
#     if "int" in str(dtype):
#         dtype_map[col] = "int64"
#     elif "float" in str(dtype):
#         dtype_map[col] = "float64"
#     elif "datetime" in str(dtype):
#         dtype_map[col] = "datetime64[us]"
#     else:
#         dtype_map[col] = "str"
# for k, v in dtype_map.items():
#     print(f'    "{k}": "{v}",')
```
